```python
# For senegal - we are trying to get the classification result for the test set



import os
import torch
import pandas as pd
import transformers

# Load AG test data (assuming: column 0 = label, column 1 = text)
df = pd.read_csv("/AG/Approaches/ag/ag_test.tsv", sep='\t', header=None, names=["label", "text"])
df['text_to_classify'] = df['text']

# Define the keyword manually (since self.keyword doesn't exist here)
search_keyword = "Senegal AND (Nutrient OR Fertilization OR Fertilizer OR Rates OR Doses OR Nitrogen OR Phosphorus OR Potassium) AND Yield."

# Initialize the model pipeline
model_id = "meta-llama/Llama-2-7b-chat-hf"
os.environ["CUDA_VISIBLE_DEVICES"] = "2"
device = "cuda" if torch.cuda.is_available() else "cpu"

pipeline = transformers.pipeline(
    "text-generation",
    model=model_id,
    model_kwargs={"torch_dtype": torch.float16},
    device=device,
)

# Prompt creation
def create_prompt(text):
    messages = [
        {"role": "system", "content": "You are an expert in agriculture and an excellent text classifier."},
        {"role": "user", "content": f'''You are given a search equation and a text. Your task is to determine if the text contains information relevant to the search equation.
Focus on finding as many keywords from the search equation as possible in the text. 
If the text contains multiple relevant keywords, respond with RELATED. If it does not contain relevant keywords or only a few, respond with UNRELATED.
Only respond with RELATED or UNRELATED no other text required.
Search Equation: {search_keyword}
Text: {text}
Result:'''}
    ]
    prompt = pipeline.tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    return prompt

terminators = [
    pipeline.tokenizer.eos_token_id,
    pipeline.tokenizer.convert_tokens_to_ids("<|eot_id|>")
]

# Classification function
def classify_text(text):
    prompt = create_prompt(text)
    outputs = pipeline(
        prompt,
        max_new_tokens=32,
        eos_token_id=terminators,
        do_sample=True,
        temperature=0.6,
        top_p=0.9,
    )
    result_text = outputs[0]["generated_text"][len(prompt):]
    return result_text.strip()

# Apply classifier to each row
results = []
for index, row in df.iterrows():
    result = classify_text(row['text_to_classify'])
    results.append(result)
    print(f"Row {index}: Predicted = {result}")

df['classification_result'] = results

# Optionally save
# df.to_csv("ag_test_classified.tsv", sep='\t', index=False)

# Optional evaluation
# Convert RELATED/UNRELATED → binary
df['pred_label'] = df['classification_result'].map(lambda x: 1 if "RELATED" in x.upper() else 0)
accuracy = (df['pred_label'] == df['label']).mean()
print(f"\nClassification Accuracy: {accuracy:.4f}")
```
